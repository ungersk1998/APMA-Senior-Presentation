```python
def agg_rent_income(df,col1,col2):
    df[col1]=df[col1].astype("float")
    df[col2]=df[col2].astype("float")
    return df.groupby("sub-burough").aggregate({col1:"mean"}).reset_index().merge(df.groupby("sub-burough").aggregate({col2:"mean"}).reset_index(),on="sub-burough")
```

```python
rent_income_average_91=agg_rent_income(clean_91,"rent_91","income_91")
rent_income_average_93=agg_rent_income(clean_93,"rent_93","income_93")
rent_income_average_96=agg_rent_income(clean_96,"rent_96","income_96")
rent_income_average_99=agg_rent_income(clean_99,"rent_99","income_99")
```
